
ini cuma catetan

* **$H_m$:** hashrate miner
* **$H_n$:** network hashrate 5 koin pow ini ambil dari jaringan masing2, bebas berapa
* **$T_b$:** waktu target dalam detik untuk menghasilkan satu blok.
* **$R$:** reward per block
* Harga ambil dari api untuk harga koin di idr
* **$W$:** konsumsi listrik dalam watt
* **$C_{kWh}$:** ambil harga pln /kwnya.
* Biaya perangkat keras kayanya ini gaperlu

```python
# kode disini dipake buat rumus ROI
# implement besok sabtu

def roi(
    hm, hn, tb, r, harga,
    watt, biaya_kwh, biaya_hw
):
    b_day = 86400 / tb
    r_day = (hm / hn) * b_day * r
    rev_day = r_day * harga
    c_day = ((watt * 24) / 1000) * biaya_kwh
    p_day = rev_day - c_day
    
    # kalo <0 berarti ga balikmodal
    if p_day <= 0:
        
        t_be = float("inf")
    else:
        t_be = biaya_hw / p_day
        
    return {
        "blok per day":     round(b_day, 4),
        "koin per day":     round(r_day, 8),
        "pendapatan kotor":  round(rev_day, 2),
        "biaya listrik":     round(c_day, 2),
        "keuntungan bersih": round(p_day, 2),
        "hari balik modal":  round(t_be, 1),
    }
```
